# Question 3 C\#

Implement function triangleArea(a,b,c) that takes as input the lengths of the 3 sides of a triangle and returns the area of the triangle. By Heron's formula, the area of a triangle with side lengths a, b, and c is square root of (s(s-a)(s-b)(s-c)), where s = (a+b+c)/2

```
namespace WorkingQuestions345.Model  // Visual Studio Syntax
{
    class Herons
    {
        //Ask for user input
        
        static void triangleArea(int side1, int side2, int side3)
        {
            double s1 = (side1 + side2 + side3)/2;

            double s2 = s1 * (s1 - side1) * (s1 - side2) * (s1 - side3);

            double s3 = Math.Sqrt(s2);
            Console.WriteLine($"The area of the triangle is {s3}");
        }
    }
}

using WorkingQuestions345.Model;

Herons cal = new Herons();
cal.triangleArea(2,2,2);
```

# Question 4 C\#

Write a program in C# to separate odd and even integers in separate arrays.

```
int[] all = new int[5];
all[0] = 25;
all[1] = 47;
all[2] = 42;
all[3] = 56;
all[4] = 32;

var even = new List<int>();
var odd = new List<int>();

foreach (int n in all)
{
    if (n % 2 == 0)
    {
        even.Add(n);
    }
    else
    {
        odd.Add(n);
    }
}

Console.WriteLine("The even numbers are:");
foreach (int num in even)
{
    Console.Write($"{num} ");
}
Console.WriteLine("\nThe odd numbers are:");
foreach (int num in odd)
{
    Console.Write($"{num} ");
}
```

Write a function inside(x,y,x1,y1,x2,y2) that returns True or False depending on whether the point (x,y) lies in the rectangle with lower left corner (x1,y1) and upper right corner (x2,y2).

```
//Visual Studio Syntax
namespace WorkingQuestion5.Model
{
    internal class rectangle
    {
        int x, y, x1, y1, x2, y2; //x,y is the point to locate, x1,y1 is the lower left of the rect, x2,y2 is the upper right
        public void inside(int x,int y)
        {
            int[,] coords = new int[3, 2] { { x, y }, { 0, 0 }, { 2, 3 } };
            x = coords[0, 0];
            y = coords[0, 1];
            x1 = coords[1, 0];
            y1 = coords[1, 1];
            x2 = coords[2, 0];
            y2 = coords[2, 1];
            if (x > x1 && x < x2)
            {
                if (y > y1 && y < y2)
                {
                    Console.WriteLine("The point falls within the rectangle");
                }
                else
                {
                    Console.WriteLine("The point is outside of the rectangle");
                }
            }
            else
            {
                Console.WriteLine("The point is outside the rectangle");
            }
        }

    }
}

using WorkingQuestion5.Model;

rectangle point = new rectangle();

point.inside(1, 1);
point.inside(-1, -1);
```
