# Question 3 C*sharp*

Implement function triangleArea(a,b,c) that takes as input the lengths of the 3 sides of a triangle and returns the area of the triangle. By Heron's formula, the area of a triangle with side lengths a, b, and c is square root of (s(s-a)(s-b)(s-c)), where s = (a+b+c)/2

```
// Visual Studio Syntax
namespace WorkingQuestions345.Model
{
    class Herons
    {
        //Ask for user input
        
        static void triangleArea(int side1, int side2, int side3)
        {
            double s1 = (side1 + side2 + side3)/2;

            double s2 = s1 * (s1 - side1) * (s1 - side2) * (s1 - side3);

            double s3 = Math.Sqrt(s2);
            Console.WriteLine($"The area of the triangle is {s3}");
        }
    }
}

using WorkingQuestions345.Model;

Herons cal = new Herons();
cal.triangleArea(2,2,2);
```

# Question 4 C*sharp*

Write a program in C# to separate odd and even integers in separate arrays.

```
int[] all = new int[5];
all[0] = 25;
all[1] = 47;
all[2] = 42;
all[3] = 56;
all[4] = 32;

var even = new List<int>();
var odd = new List<int>();

foreach (int n in all)
{
    if (n % 2 == 0)
    {
        even.Add(n);
    }
    else
    {
        odd.Add(n);
    }
}

Console.WriteLine("The even numbers are:");
foreach (int num in even)
{
    Console.Write($"{num} ");
}
Console.WriteLine("\nThe odd numbers are:");
foreach (int num in odd)
{
    Console.Write($"{num} ");
}
```

# Question 6 Python

You can turn a word into pig-Latin using the following two rules (simplified):
    1. If the word starts with a consonant, move that letter to the end and append 'ay'. For example, 'happy' becomes 'appyhay' and 'pencil' becomes 'encilpay'.
    2. If the word starts with a vowel, simply append 'way' to the end of the word. For example, 'enter' becomes 'enterway' and 'other' becomes 'otherway'. For our purposes, there are 5 vowels: a, e, i, o, u (so we count y as a consonant).
    
Write a function pig() that takes a word (i.e., a string) as input and returns its pig-Latin form. Your function should still work if the input word contains upper case characters. Your output should always be lower case however.

```
//Python 3 syntax
def pig(word):
    word = word.lower()
    if word[0] in ('a', 'e', 'i', 'o' 'u'):
        print(word+'way')
    else:
        print(word[1:]+word[0]+'ay')

pig('Robyn')
pig('Ethan')
```

and then converts and returns the amount in US dollars

```
//Python 3 syntax
print('This program will convert your currency into USD')
print('The currency types to choose from are: AUD, CHF, CNY, DKK, EUR,')
print('GBP, HKD, INR, JPY, MXN, MYR, NOK, NZD, PHP, SEK, SGD, THB')
print('')

currencytype = input('Please enter a currency type: ')
dollaramount = eval(input('Please enter a dollar amount: '))
print('')

def curconv(curtype, amount):
    infile = open('currencies.txt', 'r')
    conversion = (infile.read())
    lst = conversion.split('\n')
    for i in lst:
        sublst = i.split('\t')
        if sublst[0] == curtype:
            x = float(sublst[1])
            conv = x * amount
            print(amount,curtype,'=',conv,'USD')
            break

curconv(currencytype, dollaramount)
```

# Question 10 Python

Encryption is the process of hiding the meaning of a text by substituting letters in the message with other letters, according to some system. If the process is successful, no one but the intended recipient can understand the encrypted message. Cryptanalysis referes to attempts to undo the encryption, even if some details of the encryption are unknown (for example, if an encrypted message has been intercepted). The first step of cryptanalysis is often to build up a table of letter frequencies in the encrypted text. Assume that the string letters is already defined as 'abcdefghijklmnopqrstuvwxyz'. Write a function called frequencies() that takes a string as its only parameter, and returns a list of integers, showing the number of times each character appears in the text. Your function may ignore any characters that are not in letters.

```
//Python 3 syntax
def frequencies(word):
    letters = 'abcdefghijklmnopqrstuvwxyz'
    lst = []
    wlst = []
    clst = []
    for w in word:
        wlst.append(w)
    for l in letters:
        lst.append(l)
    for n in lst:
        if n in wlst:
            x = wlst.count(n)
            clst.append(x)
        else:
            clst.append(0)
            
    print('The encryption of',word,'is: ')
    print(clst)
    print('')

frequencies('The quick red fox got bored and went home.')
frequencies('apple')
```
